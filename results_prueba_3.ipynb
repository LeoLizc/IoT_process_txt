# Estad칤sticas de Resultados

El prop칩sito de este an치lisis es demostrar si estas se encuentran afectadas significativamente por la distancia o no. Las variables relevantes en el dataset para determinar esta relaci칩n son `my_sto`, `sto`, `snr` y `cfo`. El entorno de pruebas fue dividido en dos datasets. Uno con los datos a 4Mbps y otro a 8Mbps, con la columna `distancia` como el valor sobre el que se har치 la agrupaci칩n.

Se realizaron pruebas Shapiro-Wilk para comprobar normalidad junto con Levene para comprobar su homocedasticidad. Para el dataset de 8Mpbs, `snr` y `cfo` presentaron varianzas desiguales con normalidad, por lo que se aplic칩 Anova Welch [1]. El resto de datos no cumplieron ninguno de estos supuestos; Anova y Anova Welch fueron descartados para ellos. Por ende, se opt칩 por utilizar Kruskal-Wallis, adecuada para datos desbalanceados y con distribuci칩n no normal [2].

[1] Laerd Statistics, One-way ANOVA using SPSS Statistics. [Online]. Available: https://statistics.laerd.com/spss-tutorials/one-way-anova-using-spss-statistics.php.

[2] J. H. McDonald, Handbook of Biological Statistics, 3rd ed. Baltimore, MD: Sparky House Publishing, 2014. [Online]. Available: http://www.biostathandbook.com/kruskalwallis.html


### Pruebas de Normalidad y Homocedasticidad

```python
metricas = ['sto', 'snr', 'cfo', 'k_hat', 'k_hat2', 'muestras', 'muestras_2']
resultados_2M_ = []
resultados_8M_ = []
```

```python
df_completo_2M_ = pd.concat([df_2M60M_SF7, df_2M120M_SF7], ignore_index=True)
df_completo_2M_['distancia'] = df_completo_2M_['prueba'].apply(extraer_distancia)
df_completo_2M_['distancia'] = df_completo_2M_['distancia'].astype(float)
df_completo_2M_['sf'] = 'SF7'

df_completo_8M_SF7 = pd.concat([df_8M60M_SF7, df_8M120M_SF7], ignore_index=True)
df_completo_8M_SF7['distancia'] = df_completo_8M_SF7['prueba'].apply(extraer_distancia)
df_completo_8M_SF7['distancia'] = df_completo_8M_SF7['distancia'].astype(float)
df_completo_8M_SF7['sf'] = 'SF7'

df_completo_8M_SF12 = pd.concat([df_8M60M_SF12, df_8M120M_SF12], ignore_index=True)
df_completo_8M_SF12['distancia'] = df_completo_8M_SF12['prueba'].apply(extraer_distancia)
df_completo_8M_SF12['distancia'] = df_completo_8M_SF12['distancia'].astype(float)
df_completo_8M_SF12['sf'] = 'SF12'

df_completo_8M_ = pd.concat([df_completo_8M_SF7, df_completo_8M_SF12], ignore_index=True)
```

2MM

```python
for metrica in metricas:
    if metrica in df_completo_2M_.columns:
        resultado = comparar_metricas_por_distancia(df_completo_2M_, metrica, columna_grupo='distancia')
        resultados_2M_.append(resultado)
        
pd.DataFrame(resultados_2M_)
```

8MM

```python
for metrica in metricas:
    if metrica in df_completo_8M_.columns:
        resultado = comparar_metricas_por_distancia(df_completo_8M_, metrica, columna_grupo='distancia')
        resultados_8M_.append(resultado)
pd.DataFrame(resultados_8M_)
```
